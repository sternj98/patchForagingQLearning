```python
import numpy as np
from matplotlib import pyplot as plt
import pandas as pd
import seaborn as sns
from scipy.stats import pearsonr,linregress
from scipy.optimize import least_squares
```

## 1.c.iii)

```python
x = np.linspace(-3,3,100)
ux = 2*x + 2 * (x**2 - 1)
gx = 2 * x

plt.figure()
plt.plot(x,ux,label = 'ux')
plt.plot(x,gx,label = 'gx')
plt.legend()
```

## 3.a.ii)

```python
theta = 1 
x = np.linspace(-3,3,100)
mu = np.exp(theta * x) / (1 + np.exp(theta * x))
mu_prime = theta * mu * (1 - mu)
lam = np.log(mu) - np.log(1 - mu)

plt.figure()
plt.plot(x,mu,label = 'mu')
plt.plot(x,mu_prime,label = 'mu prime')
plt.legend()
plt.figure()
plt.plot(x,lam,label = 'lambda')
plt.legend()
```

```python
#a)#
data = pd.read_csv('scores.csv')
sns.scatterplot(x = 'midterm', y = 'final',data = data)
corr = pearsonr(data['midterm'],data['final'])
print('Sample Correlation: ',corr[0])
```

```python
#b)#
a1,b1 = linregress(data['midterm'],data['final'])[0:2]
plt.figure()
sns.scatterplot(x = 'midterm', y = 'final',data = data)
print('slope = ',a1)
print('intercept = ',b1)
x = np.linspace(0,100,100)
y_hat = a1 * x + b1
plt.plot(y_hat,'r')
```
